# Analysis for confirmed,recovered and total deaths in one graph

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
url_confirmed = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv"
url_recovered = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv"
url_death = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv"
```

```python
df_confirmed = pd.read_csv(url_confirmed)
df_recovered = pd.read_csv(url_recovered)
df_death = pd.read_csv(url_death)
```

```python
df_confirmed
```

```python
df_recovered
```

```python
df_death
```

```python
def df_new(df):
    drop_col = ["Province/State","Lat","Long"]
    df = df.drop(drop_col,axis=1)
    return df.groupby("Country/Region",as_index=False).agg(sum)
```

```python
df_con_new = df_new(df_confirmed)
df_recov_new = df_new(df_recovered)
df_death_new = df_new(df_death)
```

# for confirmed

```python
df_con_new
```

```python
df_con_new_one = df_con_new.loc[df_con_new["Country/Region"] == "India"].drop("Country/Region",axis=1).T.reset_index()
```

```python
df_con_new_one.columns = ["Dates","Confirmed"]
```

```python
df_con_new_one
```

# for recovered

```python
df_recov_new_one = df_recov_new.loc[df_recov_new["Country/Region"] == "India"].drop("Country/Region",axis=1).T.reset_index()
```

```python
df_recov_new_one.columns = ["Dates","Recovered"]
```

```python
df_recov_new_one
```

# for death

```python
df_death_new_one = df_death_new.loc[df_death_new["Country/Region"] == "India"].drop("Country/Region",axis=1).T.reset_index()
```

```python
df_death_new_one.columns = ["Dates","Deaths"]
```

```python
df_death_new_one
```

```python
df_all = pd.merge(df_con_new_one,df_recov_new_one,on="Dates",how="left")
```

```python
df_all_one = pd.merge(df_all,df_death_new_one,on="Dates",how="left")
```

```python
df_all_one["Dates"] = pd.to_datetime(df_all_one["Dates"])
```

```python
df_all_one.set_index("Dates",inplace=True)
```

```python
df_all_one.plot()
plt.title("Plot for India")
plt.show()
```
