# Plotting Confirmed,Recovered,and Deaths in one graph for a one country

```python
def filter_for_country(df,country_name):
    df_new = df.loc[df["Country/Region"].str.lower() == country_name].drop("Country/Region",axis=1).T
    return df_new
```

```python
def get_all_in_one(list_df,country_name):
    country_name = country_name.lower()
        
    list_df_crd = [filter_for_country(x,country_name) for x in list_df]
    comb_df = pd.concat(list_df_crd, axis = 1).reset_index()
    comb_df.columns = ["Date","Confirmed","Recovered","Death"]
    comb_df["Active"] = comb_df["Confirmed"] - comb_df["Recovered"] - comb_df["Death"]
    comb_df['Date'] = pd.to_datetime(comb_df['Date'])
    comb_df.set_index("Date",inplace=True)
    comb_df.plot()
    plt.ylabel("Total Cases")
    plt.title("Plot For {}".format(country_name.upper()))
    plt.show()
```

```python
get_all_in_one([df_con_new,df_recov_new,df_death_new],"germany")
```

## For a particular country, C,R,D,A,or C_Incr,R_Incr and % of each in Dict format using Function

```python
def get_percent_inc(df,col) :
    ser = np.round(100 * df[col+"_inc"]/(df[col] - df[col+"_inc"]),2)
    ser = ser.replace(np.nan,0)
    ser = ser.astype(str) + "%"
    return ser
```

```python
def filter_for(df,country_name):
    df_new = df.loc[df["Country/Region"].str.lower() == country_name].drop("Country/Region",axis=1).T
    return df_new
```

```python
def get_extra_data(comb_df):
    
    comb_df_diff = comb_df.diff(1)
    comb_df_diff.columns = ["Confirmed_inc","Recovered_inc","Deaths_inc","Active_inc"]
        
    df_total_diff = pd.concat([comb_df,comb_df_diff],axis=1)
    df_total_diff["Confirmed_inc_%"] = get_percent_inc(df_total_diff,"Confirmed")
    df_total_diff["Recovered_inc_%"] = get_percent_inc(df_total_diff,"Recovered")
    df_total_diff["Deaths_inc_%"] = get_percent_inc(df_total_diff,"Deaths")
    df_total_diff["Active_inc_%"] = get_percent_inc(df_total_diff,"Active")
    
#     df_total_diff["Confirmed_inc_%"] = np.round(100 * df_total_diff["Confirmed_inc"]/(df_total_diff["Confirmed"] - df_total_diff["Confirmed_inc"]),2)
#     df_total_diff["Recovered_inc_%"] = np.round(100 * df_total_diff["Recovered_inc"]/(df_total_diff["Recovered"] - df_total_diff["Recovered_inc"]),2)
#     df_total_diff["Deaths_inc_%"] = np.round(100 * df_total_diff["Deaths_inc"]/(df_total_diff["Deaths"] - df_total_diff["Deaths_inc"]),2)
#     df_total_diff["Active_inc_%"] = np.round(100 * df_total_diff["Active_inc"]/(df_total_diff["Active"] - df_total_diff["Active_inc"]),2)
    
    df_total_diff_new = df_total_diff.replace(np.nan,0)
    df_total_diff_new.reset_index(inplace = True)
    df_total_diff_new['Date'] = df_total_diff_new['Date'].astype(str)
    return df_total_diff_new
#     return df_total_diff
```

```python
def get_all_data(list_df,country_name):
    country_name = country_name.lower()
        
    list_df_crd = [filter_for(x,country_name) for x in list_df]
    comb_df = pd.concat(list_df_crd, axis = 1).reset_index()
    comb_df.columns = ["Date","Confirmed","Recovered","Deaths"]
    comb_df["Active"] = comb_df["Confirmed"] - comb_df["Recovered"] - comb_df["Deaths"]
    comb_df['Date'] = pd.to_datetime(comb_df['Date'])
    comb_df.set_index("Date",inplace=True)
    return comb_df
```

```python
df_combined = get_all_data([df_con_new,df_recov_new,df_death_new],"india")
```
