# Analysis For Death Rate For Each Country

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
url = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv"
```

```python
df = pd.read_csv(url)
```

```python
df.head()
```

```python
drop_col = ["Province/State","Lat","Long"]
```

```python
df_new = df.drop(drop_col,axis=1)
```

```python
df_new_grouped = df_new.groupby("Country/Region",as_index=False).agg(sum)
```

```python
df_new_grouped
```

# plotting graph of a particular country using function

```python
def plot_for_death(country_name) :
    country_name = country_name.lower()
    df_one = df_new_grouped.loc[df_new_grouped["Country/Region"].str.lower() == country_name].drop("Country/Region",axis=1).T
    df_new_col = df_one.reset_index()
    df_new_col.columns = ["Dates","Total_Cases"]
    df_new_col["Dates"] = pd.to_datetime(df_new_col["Dates"])
    df_new_col.plot(kind="line",x="Dates",y="Total_Cases")
    plt.ylabel
    plt.title("Death Rate for {}".format(country_name.upper()))
    plt.show()
```

```python
plot_for_death("india")
```

# Top 10 Countries which has most deaths yet

```python
df_new_grouped
```

```python
def death_for_top_10():
    df_top_10 = df_new_grouped[["Country/Region","6/7/20"]].sort_values("6/7/20",ascending=False).head(10)
    df_top_10.plot(kind="bar",x="Country/Region",y="6/7/20")
    plt.ylabel("Total Deaths")
    plt.show()
```

```python
death_for_top_10()
```

# Increase in Daily Cases for each country

```python
def daily_death_rate(country_name):
#     country_name = country_name.lower()
    df_new_tr = df_new_grouped.set_index("Country/Region").T
    df_new_tr = df_new_tr.diff().replace(np.nan,0.0)
    df_new_tr[country_name].plot(x="line")
    plt.xlabel("Date")
    plt.ylabel("Total Deaths")
    plt.title("daily death rate for {}".format(country_name.upper()))
    plt.show()
```

```python
daily_death_rate("India")
```
