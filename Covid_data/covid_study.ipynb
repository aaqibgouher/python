```python
import pandas as pd
import matplotlib.pyplot as pt
import seaborn as sns
```

```python
url = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv"
```

# Reading Raw Data

```python
df = pd.read_csv(url)
```

```python
df
```

```python
df.head()
```

# Selecting Required Columns

```python
drop_cols = ['Province/State','Lat','Long']
```

```python
req_cols = [x for x in df.columns if x not in drop_cols]
```

```python
df_new = df.loc[:,req_cols]
```

```python
df_new
```

# Particular Country

```python
country = 'India'
```

```python
df_new['Country/Region']==country
```

```python
df_new.loc[df_new['Country/Region']==country]
```

```python
df_india = df_new.loc[df_new['Country/Region']==country].drop('Country/Region', axis = 1).T

df_india.reset_index(inplace = True)

df_india.columns = ['Date','total_cases']
```

```python
df_india
```

```python
df_india['Date'] = pd.to_datetime(df_india['Date'])
```

```python
df_india.info()
```

```python
df_india.plot(kind = 'scatter', x = 'Date', y= 'total_cases')
pt.xticks(rotation = 30)
pt.title("For India")
pt.show()
```

# Function for Particular Country

```python
def plot_for_country(country_name):
    dfc = df_new.loc[df_new['Country/Region']==country_name].drop('Country/Region', axis = 1).T
    dfc.reset_index(inplace = True)
    dfc.columns = ['Date','total_cases']
    dfc['Date'] = pd.to_datetime(dfc['Date'])
    dfc.plot(kind = 'scatter', x = 'Date', y= 'total_cases')
    pt.xticks(rotation = 30)
    pt.title(f"For {country_name}")
    pt.show()
```

```python
plot_for_country("Russia")
```
