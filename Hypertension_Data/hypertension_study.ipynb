```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Data Reading

```python
df = pd.read_csv("Hypertension_data.csv")
```

# Categorical Feature

**1) prevalentHyp - Using matplotlib**

```python
df_phyp = df.prevalentHyp.value_counts()
```

```python
df_phyp
```

```python
df_phyp.plot(kind='bar')
plt.xlabel('dffssd')
```

**2) prevalentHyp - Using seaborn**

```python
import seaborn as sns
```

```python
sns.countplot(x = df.prevalentHyp)
```

**3) education**

```python
sns.countplot(x=df.education)
```

**4)Current Smokers**

```python
sns.countplot(x=df.currentSmoker)
```

**5) age**

```python
df.age.plot(kind = 'hist', bins = 30)
```

```python
sns.distplot(df.age)
```
