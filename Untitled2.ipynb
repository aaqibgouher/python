```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
url = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv"
```

```python
df = pd.read_csv(url)
```

```python
df
```

```python
drop_col = ["Province/State","Lat","Long"]
```

```python
df_new = df.drop(drop_col,axis=1)
```

```python
df_new.head()
```

```python
df_df = df_new.loc[df_new["Country/Region"] == "China"]
# df_df.set_index("Country/Region",inplace=True)
df_df
```

```python
df_df = df_df.groupby("Country/Region").agg(sum)
df_c = df_df.T.reset_index()
# df_c.columns.tolist()
df_c.columns = ["Dates","Total_Cases"]
df_c["Dates"] = pd.to_datetime(df_c["Dates"])
df_c.plot(kind="line",x="Dates",y="Total_Cases")
plt.ylabel("Total_Cases")
plt.show()
```
