# Options ans Customisation

*suppose we have a df,which has lots of rows and columns . so when we print our df , it will not show all of the rows and columns. it will show some from the first and some from the last. lets take an eg. suppose for rows , it has given 20 datas from the first and then some ..... then last 20 datas from tail. so we can change it using some of the options in pandas

# It is of 5 types

```python
import pandas as pd
import numpy as np
```

```python
url = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv"
```

```python
df = pd.read_csv(url)
```

# 1. get_option(p1,p2)

here as you can see that its showing first 5 and last 5 of the df.

```python
df
```

using this, we can find , how much rows, it will show to us

```python
pd.get_option("display.max_rows")
```

using this ,we can change the size by passing 2nd para as in the funct. None means if we have infinite number of rows ,then also it will show all of these.

```python
pd.set_option("display.max_rows",None)
```

```python
df
```

suppose we have changed the no, of rows, and we want to change it again , so we have to done this.

```python
pd.reset_option("display.max_rows",None)
```

```python
df
```

for column, same task we have to performed

```python
pd.get_option("display.max_columns")
```

```python
pd.set_option("display.max_columns",None)
```

```python
df
```

```python
pd.reset_option("display.max_columns")
```

```python
df
```

describe_option prints the description of the argument.

```python
pd.describe_option("display.max_columns")
```

```python
df
```

The Columns , which are in float , it will take into 2 decimals

```python
pd.set_option("display.precision",2)
```
