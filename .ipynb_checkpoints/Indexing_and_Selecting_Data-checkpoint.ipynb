Using loc , we can selet the row as well as column datas. here we have selected row (0 - 4) and a specific columns. It will give us a Series.

```python
df.loc[0:4,"Lat"]
```

How to select a mupltiple columns. It will give us a df.

```python
df.loc[0:4,["Lat","Long"]]
```

For specific Rows ans Cols : It will return a df

```python
df.loc[[0,2,4,6,8],["Lat","Long"]]
```

For Specific rows but all cols :

```python
df.loc[0:5]
```

Use of ILOC() : used for integers only

```python
df.iloc[1] b
```

```python
df.iloc[[1,3,5]]      # will select only that rows
```

```python
df.iloc[[1,3,5],[0,1,2,3]]   # will choose 1,3,5 rows and the columns 0,1,2,3 .since it is used only for int thats why we have to use int index
```

```python
df.iloc[0:5]     # will select rows from 0 to 5 (exclusive). df.iloc[0:5,:] alter
```

```python
df.iloc[0:5,0:4]      # will select rows from 0 to 4 and cols from 0 to 3
```
